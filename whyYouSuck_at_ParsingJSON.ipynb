# `This will take uneven length columns and use pd.explode`

+ `pd.explode` takes a column of lists and removes the list and creates more rows, with whatever is inside of the lists.

This example is different because you are using:
+ More than one column
+ and they are all different size lists

If they were the same length: something such as `df.apply(pd.Series.explode)`
can be used.

**But, you need to understand you are creating MORE ROWS!**
+ The rows have an index of where they were coresponding to, in regards to original data.


**If this is undesirable, then leave as lists!**
+ Because, you don't have any other options since the information in the lists are not consistent. If they were you could create new columns and match.

```python
mod_df=news_df.loc[:,['des_facet','org_facet','per_facet',
                      'geo_facet','media']]

for i in mod_df.columns:
    mod_df=mod_df.explode(i)
mod_df
# mod_df.explode('des_facet')
```

# `How do we deal with the dictionary left behind now?`

+ This is where **`json.normalize can be used`**

```python
flat_dct=pd.json_normalize(json.loads(mod_df.\
                                to_json(orient="records")))
flat_dct.head()
```
